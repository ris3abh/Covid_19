```python
import pandas as pd
import os
import shutil
import random
```

```python
## positive samples from github
file_loc = '../../chestxray/metadata.csv'
image_loc = '../../chestxray/images'
```

```python
df = pd.read_csv(file_loc)
df.shape
```

```python
target_loc = '../../Data_set/covid'
```

```python
if not os.path.exists(target_loc):
    os.mkdir(target_loc)
    print('Covid Folder Created')
```

```python
print(df["view"].head())
print(df['view'].tail())
```

```python
count = 0
for (i,row) in df.iterrows():
    if row['finding'] == 'Pneumonia/Viral/COVID-19' and row['view'] == 'PA':
        filename = row['filename']
        image_path = os.path.join(image_loc,filename)
        image_copy_path = os.path.join(target_loc,filename)
        shutil.copy2(image_path,image_copy_path)
        print('moving Image',count)
        count+=1
print(count)
```

```python
kaggle_loc = '../../chest_xray/train/NORMAL'
target2_loc = '../../Data_set/normal'
```

```python
image_names = os.listdir(kaggle_loc)
random.shuffle(image_names)
image_names
```

```python
count = 0
for i in range(196):
    image_name = image_names[i]
    image_path = os.path.join(kaggle_loc,image_name)
    target2_path = os.path.join(target2_loc,image_name)
    shutil.copy2(image_path,target2_path)
    print('copying image',count)
    count+=1
```
